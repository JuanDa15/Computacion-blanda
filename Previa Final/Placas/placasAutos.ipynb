```python

# Instalación de OpenCV
# pip install opencv-contrib-python

# Instalación de la librería de Tesseract
# pip install pytesseract

# Instalación del programa Tesseract
# Instalar el programa Tesseract.exe

# Importar librería OpenCV
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import cv2

# Importar Librería para reconocimiento de caracteres
import pytesseract

# Direccionar a programa Tesseract instalado
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'

# Se guardan las coordenadas de donde  se encuentra ubicada la placa.
placa = []

#Lectura de la imagen de la cual se extrae la placa.
#Imagenes probadas:
#   auto001.jpg
#   auto002.jpeg
#   auto003.jpg
#   fr.jpg

image = cv2.imread(r'C:\Users\Jdoo1\Documents\Computacion-blanda\Final\Placas\auto002.jpeg')
#Muestra la imagen con la que se va a trabajar
#cv2.imshow('imagen',image)

# Conversión a escala de grises
#Muestra la imagen en escala de grises
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#cv2.imshow('imagen2',gray)

# Proceso de suavizado de la imagen
gray = cv2.blur(gray,(3,3))
#Muestra la imagen con un difuminado
#cv2.imshow('imagen3',gray)

# Algoritmo de detección de bordes
canny = cv2.Canny(gray,150,200)
#Muestra la imagen resaltando todos los bordes que detecte
#cv2.imshow('imagen4',canny)

# Algoritmo para resaltar los bordes detectados
canny = cv2.dilate(canny,None,iterations=1)
#Le da una amplitud de mas pixeles a los bordes
#cv2.imshow('imagen5',canny)

# Algoritmo para unir de manera continua los contornos
# Encuentra todos los contornos presentes en la imagen
#_,cnts,_ = cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
cnts,_ = cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

# Función para dibujar todos los contornos encontrados
#cv2.drawContours(image,cnts,-1,(0,255,0),2)

#Ciclo que se encarfa de recorrer todos los contornos encontrados.
#Cada contorno encontrado es un arreglo.
for c in cnts:
    # Encuentra el área del contorno
    area = cv2.contourArea(c)
    
    # Calcula el rectángulo del contorno
    #boundingRect se encarga de retornar las coordenadas del punto superior izquierdo, ancho y alto del rectangulo
    x,y,w,h = cv2.boundingRect(c)
    #Se calcula el valor de epsilon para la funcion de aproximaxion.
    epsilon = 0.09*cv2.arcLength(c,True)
    approx = cv2.approxPolyDP(c,epsilon,True)

    # Si tiene 4 vértices y un área aproximada de 9000, parece una placa
    #Se debe ajustar el area segun el tamaño de las placas que queramos encontrar
    if len(approx)==4 and area>9000:
        # Imprime el área
        print('area=',area)
    
        # Calcula la relación anchura-altura
        #cv2.drawContours(image,[approx],0,(0,255,0),3)
        aspect_ratio = float(w)/h
    
        # Si la relación es 2.4, el rectángulo es muy parecido a la placa
        # Se tiene que cambiar la relación de aspecto segun la imagen ya que todas
        # las placas no tienen la misma relación de aspecto
        #auto003.jpg ----> 2.0
        #auto002.jpeg ----> 1.7
        #auto001.jpg ----> 3.1
        #fr.jpg ------> 2.9
        
        #Se imprime la relacion de aspecto
        
        #Relaciones de aspecto mayores a las de la imagen no dejan que el programa continua con normalidad
        #Por ende con cada imagen hay que verificar la relación de aspecto
        print(aspect_ratio)
        
        if aspect_ratio>1.7:
    
            # En las siguientes instrucciones se imprimen los datos de la placa 
            # y se la muestra en una ventana
            placa = gray[y:y+h,x:x+w]
      
            # Se extrae el texto asociado a la placa utilizando Tesseract
            text = pytesseract.image_to_string(placa,config='--psm 11')
    
            # Se imprime la texto de la placa encontrada
            print('PLACA: ',text)
    
            # Se muestra una imagen de la placa
            cv2.imshow('PLACA',placa)
    
            # Se agrega un texto a la imagen, indicando los datos encontrados
            cv2.moveWindow('PLACA',780,10)
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3)
            cv2.putText(image,text,(x-20,y-10),1,1.2,(0,255,0),3)

# Se muestran los datos obtenidos
cv2.imshow('Image',image)
cv2.moveWindow('Image',45,10)
cv2.waitKey(0)
```
